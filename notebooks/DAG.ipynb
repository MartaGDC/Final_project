**A socioecological measurement of homophobia for all countries and its public health impact**
2018



The availability and reliability of data on stigma and discrimination related to gender identity and sexual orientation and gender identity is uneven from country to country. Data on HIV and on gay men and other MSM were obtained from AIDSinfo 20 an online repository of global AIDS data compiled and maintained by UNAIDS, as well as from scientific articles and grey literature.

Economic and development indicators such as:
- gross domestic product
- health expenditures
- proportion of seats held by women in parliaments
- gender inequality
- mean years of schooling
- life expectancy

were drawn from the IMF World Economic Outlook, the 2015 Human Development Report and the World Population Prospects published by the Department of Economic and Social Affairs of the United Nations Population Division.

We used the individual data from the World Values Survey which we aggregated at country level. - *requires login*

The information on laws and legislation comes from the International Lesbian, Gay, Bisexual, Trans and Intersex Association(ILGA).

The rule of law index comes from the World Justice Project and the Cantril index from the Gallup World Poll. We used the data on homophobia made available by the Pew ResearchCenter.


```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
dag = nx.DiGraph()

# Add the nodes with position attributes
dag.add_node("\nInstitutional discrimination", exposure=True, pos=(0.296, -0.243))
dag.add_node("Internalized homophobia", latent=True, pos=(0.648, -0.557))
dag.add_node("Social\ndiscrimination\n\n", exposure=True, pos=(1.111, 0.311))
dag.add_node("Characteristics\nof the country", exposure=True, pos=(-0.346, 0.188))
dag.add_node("Sentiment", outcome=True, pos=(0.591, -1.732))
node_positions = {node: attributes["pos"] for node, attributes in dag.nodes(data=True)}

# Add arrows
dag.add_edges_from([
    ("\nInstitutional discrimination", "Internalized homophobia"),
    ("\nInstitutional discrimination", "Social\ndiscrimination\n\n"),
    ("\nInstitutional discrimination", "Sentiment"),
    ("Internalized homophobia", "Sentiment"),
    ("Social\ndiscrimination\n\n", "Internalized homophobia"),
    ("Social\ndiscrimination\n\n", "Sentiment"),
    ("Characteristics\nof the country", "\nInstitutional discrimination"),
    ("Characteristics\nof the country", "Social\ndiscrimination\n\n"),
    ("Characteristics\nof the country", "Sentiment"),
])

nx.draw(dag, pos=node_positions, with_labels=True, node_size=700, node_color="lightblue", font_size=8, font_color="black", font_weight="bold", arrowsize=20)
plt.show()
```
